```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def model(X,t):
    a = X[0]
    b = X[1]
    c = X[2]
    d = X[3]
    
    dadt= (-a + c*d)
    dbdt= a
    dcdt = (2*b) - (c*d)
    dddt = b - (c*d)
    
    dXdt = [dadt, dbdt, dcdt, dddt]
    return dXdt

X0 = [100,0,0,0]
t = np.linspace(0,20)

X = odeint(model,X0,t)
plt.plot(t,X)
plt.show()
    
    
```
