```python
import numpy as np
import matplotlib.pyplot as plt

# For sanger sequencing

L = 800 #read length
N = 4600000 #Genome size

xvalues = [] #For holding values after calculation

probabilities = list(range(0,101)) #List from 0 to 101

for i in range(0,101):
    probabilities[i] = probabilities[i]/100 #Converts integers to percentages
    
probabilities.pop(0) #Removes 0 in list (can't log(0))

for i in probabilities:
    xvalues.append(np.log(i) / np.log(1-(L/N))) #Calculation of number of reads for (y) probability
    
probabilities.reverse() #0 to 1 instead of 1 to 0

plt.title("Probability as a Function of Coverage for Sanger")
plt.xlabel("Coverage")
plt.ylabel("Probability")
plt.plot(xvalues,probabilities)
plt.show()

```
