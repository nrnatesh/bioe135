```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

k1 = 1
k2 = 1
k3 = 1
k4 = 1

def model(X,t):
    a = X[0]
    b = X[1]
    c = X[2]
    d = X[3]
    
    dadt= (-k1 * a) + (k4*c*d)
    dbdt= k1*a - (k2*b) - (k3*b)
    dcdt = (2*k2*b) - (k4*c*d)
    dddt = k3*b - (k4*c*d)
    
    dXdt = [dadt, dbdt, dcdt, dddt]
    return dXdt

X0 = [100,0,0,0]
t = np.linspace(0,5)

X = odeint(model,X0,t)

plt.xlabel('Time steps')
plt.ylabel('[Species X]')
plt.title('Time Course of System')
plt.plot(t,X[:,0],label = 'A', color = 'r')
plt.plot(t,X[:,1],label = 'B', color = 'c')
plt.plot(t,X[:,2],label = 'C', color = 'b')
plt.plot(t,X[:,3],label = 'A', color = 'y')
plt.legend()
plt.show()

    
```

This is a paper by Milker et. al attempting to model the kinetics of two redox cascades. Through this, they found that the availability of redox cofactors is significant in optimizing the models, as are their fluxes upon simulation. The amount and stability of all enzymes involved are also important.
They sought to estimate enzyme concentrations in two artificial redox cascades in order to identify the rate-determining step. They also wanted to identify possible extrinsic reasons for the reduced velocity within the cascade. From this, sought to propose an optimization strategy to overcome this limitation. The reason for a model was to be able to quantify the performance of the pathway, without actually knowing what the rate-limiting catalytic step was.

## Assumptions:
Rate laws were derived from MM kinetic model. Concentrations of cofactors NADP+ and NADPH would remain constant, given e. coli’s ability actively regulate redox balance for homeostasis. Also assumed all enzymes would suffice under MM kinetics, possibly with inhibition (EtOH in the solvent). Assumed the biocatalysts would not degrade over the time-course.

## Limitation:
A limitation is that their initial parameters were estimated by experimental work, but they admitted themselves that this work yielded “insufficient precision and sampling frequency” meaning their values varied greatly. Therefore, the time-course of the system could actually vary greatly since the initial parameters are suspicious.
